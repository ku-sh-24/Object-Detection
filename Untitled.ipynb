```python
import os
import random
import shutil

# Define paths for your image directories
base_path = 'dataset/Data/images'  # The path where your images are stored
output_path = 'Final'  # Where you want to store the train, valid, test sets
class_names_file = 'dataset/classes.txt'  # Path to your TXT file containing the class names
labels_path = 'dataset/Data/labels'   # Path where your label .txt files are stored


# Define class names directly in the code (as a dictionary)
classes = {
    0: 'car',
    1: 'threewheel',
    2: 'bus',
    3: 'truck',
    4: 'motorbike',
    5: 'van'
}

# Function to parse the label files and associate images with class indices
def parse_labels(classes):
    labels = {idx: [] for idx in classes.keys()}  # Prepare lists for each class index

    # Iterate through all image files (assuming .jpg, .jpeg, .png images)
    for image_name in os.listdir(base_path):
        if image_name.endswith(('.jpg', '.jpeg')):  # Add other formats if necessary
            txt_file = os.path.splitext(image_name)[0] + '.txt'  # The corresponding label file
            txt_file_path = os.path.join(labels_path, txt_file)

            # If the corresponding txt file exists, read its content
            if os.path.exists(txt_file_path):
                with open(txt_file_path, 'r') as label_file:
                    # Read each line in the label file
                    lines = label_file.readlines()
                    for line in lines:
                        parts = line.strip().split()  # Split by whitespace
                        class_index = int(parts[0])  # Class index is the first value in the line

                        # Debug: print class index and class name
                        if class_index in classes:
                            class_name = classes[class_index]
#                             print(f"Image: {image_name}, Class Index: {class_index}, Class Name: {class_name}")
                            labels[class_index].append(image_name)  # Add image to the corresponding class
#                         else:
#                             print(f"Warning: Class index {class_index} is not valid. Skipping image {image_name}.")
            else:
                print(f"Warning: No label file found for {image_name} in {labels_path}")
    
    return labels

# Parse the label files and organize images by class
labels_dict = parse_labels(classes)

# Create output directories if they don't exist
os.makedirs(os.path.join(output_path, 'train'), exist_ok=True)
os.makedirs(os.path.join(output_path, 'valid'), exist_ok=True)
os.makedirs(os.path.join(output_path, 'test'), exist_ok=True)

# Function to split and copy images into respective folders
def split_and_copy_images():
    for class_index, images in labels_dict.items():
        class_name = classes[class_index]  # Get the class name for the current index
#         print(f"Processing class: {class_name} (Index {class_index}), {len(images)} images found.")
        
        # Ensure there are at least 100 images for the current class
#         if len(images) < 70:
#     train_images = images
# else:
#     selected_images = random.sample(images, 100)
#     train_images = selected_images[:70]
        if len(images) < 100:
            print(f"Warning: {class_name} has less than 100 images. Only {len(images)} images available.")
            selected_images = images  # Select all available images
        else:
            # Randomly sample 100 images
            selected_images = random.sample(images, 100)

        # Split selected images into train, valid, test
        train_images = selected_images[:70]
        valid_images = selected_images[70:90]
        test_images = selected_images[90:]

        # Define destination directories for each set
        train_dir = os.path.join(output_path, 'train', class_name)
        valid_dir = os.path.join(output_path, 'valid', class_name)
        test_dir = os.path.join(output_path, 'test', class_name)

        # Create directories for each class in train, valid, test
        os.makedirs(train_dir, exist_ok=True)
        os.makedirs(valid_dir, exist_ok=True)
        os.makedirs(test_dir, exist_ok=True)

        # Copy images into the respective folders
        for img in train_images:
            img_path = os.path.join(base_path, img)
            if os.path.exists(img_path):
                shutil.copy(img_path, os.path.join(train_dir, img))
#             else:
#                 print(f"Warning: Image {img} not found in {base_path}")

        for img in valid_images:
            img_path = os.path.join(base_path, img)
            if os.path.exists(img_path):
                shutil.copy(img_path, os.path.join(valid_dir, img))
#             else:
#                 print(f"Warning: Image {img} not found in {base_path}")

        for img in test_images:
            img_path = os.path.join(base_path, img)
            if os.path.exists(img_path):
                shutil.copy(img_path, os.path.join(test_dir, img))
#             else:
#                 print(f"Warning: Image {img} not found in {base_path}")

#         print(f"Processed {class_name}: {len(train_images)} train, {len(valid_images)} valid, {len(test_images)} test images.")

# Run the function to split and copy images
split_and_copy_images()
```

```python
import os

# Function to count the number of images in each folder (train, valid, test) for each class
def count_images_in_folders(base_output_path, classes):
    counts = {'train': {}, 'valid': {}, 'test': {}}
    
    # Iterate through each set (train, valid, test)
    for split in ['train', 'valid', 'test']:
        for class_index, class_name in classes.items():
            class_folder = os.path.join(base_output_path, split, class_name)
            if os.path.exists(class_folder):
                image_count = len([f for f in os.listdir(class_folder) if f.endswith(('.jpg', '.jpeg', '.png'))])
                counts[split][class_name] = image_count
            else:
                counts[split][class_name] = 0  # If the folder doesn't exist or is empty
    
    # Print out the image counts for each class and split
    for split in ['train', 'valid', 'test']:
        print(f"\n{split.capitalize()} set image counts:")
        for class_name in classes.values():
            print(f"{class_name}: {counts[split].get(class_name, 0)} images")

    return counts
```
